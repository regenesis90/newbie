## 2. icrawler 임포트, 저장 경로 및 폴더 설정

```python
# icrawler 패키지, builtin 모듈, BingImageCrawler, BaiduImageCrawler 클래스 임포트
import icrawler
from icrawler.builtin import BingImageCrawler, BaiduImageCrawler
```

```python
# 이미지 다운로드 받을 폴더 관리 목적 : 파일 관리, 폴더 생성 및 삭제
import os
```

```python
# 저장할 디렉토리 문자열 세팅
default_dir='E:/git_github_regenesis90/Python_JupyterLab'
download_dir=os.path.join(default_dir,'downloads') #> 기본 경로에 'downloads' 폴더 생성
bing_dir=os.path.join(download_dir,'bing') #> Bing 검색엔진 이미지 저장 디렉토리
baidu_dir=os.path.join(download_dir,'baidu') #> Baidu 검색엔진 이미지 저장 디렉토리
```

```python
# bing 디렉토리가 존재하지 않을 시 : 디렉토리를 생성함
if not os.path.exists(bing_dir):
    os.makedirs(bing_dir)
```

```python
# baidu 디렉토리가 존재하지 않을 시 : 디렉토리(baidu 폴더)를 생성함.
if not os.path.exists(baidu_dir):
    os.makedirs(baidu_dir)
```

## 3. Bing Image Crawl

### 3-1. Bing Image Crawl

```python
# BingImageCrawler 클래스 인스턴스 생성 : bing_dir 디렉토리에 저장하는 BingImageCrawler
bing_crawler=BingImageCrawler(storage={'root_dir':bing_dir})
```

```python
# Crawler 객체의 crawl 메서드 호출 : 이미지 다운로드
# keyword : 검색어, max_num : 다운로드 받으려는 (최대) 이미지 개수
bing_crawler.crawl(keyword='dog', max_num=10)
```

### 3-2. 콘솔에 다운로드 받은 이미지 출력하기

* matplotlib 패키지, PIL(Pillow) 패키지 필요

```python
# matplotlib 패키지, pyplot 모듈을 plt 로 줄여 임포트
import matplotlib.pyplot as plt
```

```python
# PIL 패키지, Image 클래스 임포트
from PIL import Image
```

```python
# 이미지 파일 이름 지정:
image_file=os.path.join(bing_dir, '000001.jpg')
# 이미지 파일 오픈:
image=Image.open(image_file)
#이미지 파일을 콘솔에 출력:
plt.imshow(image)
```

## 4. Baidu Image Crawl

### 4-1. Baidu Image Crawl

```python
# BaiduImageCrawler 클래스의 객체 baidu_crawler 생성
baidu_crawler=BaiduImageCrawler(storage={'root_dir':baidu_dir},
                               parser_threads=2,
                               downloader_threads=4)
# 이미지 다운로드(keyword : 고양이, 사진 개수 : 12개, 파일 덮어쓰기 : 허용)
baidu_crawler.crawl(keyword='cat', max_num=12, overwrite=True)
```

### 4-2. 이미지를 콘솔에 출력하기

```python
# 이미지파일 이름 지정
image_file=os.path.join(baidu_dir, '000001.jpg')
# 이미지 파일 오픈
image=Image.open(image_file)
# 이미지 파일을 콘솔에 출력
plt.imshow(image)
```
