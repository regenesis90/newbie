# Boxplot

## seaborn의 sns.boxplot() 사용

```python
# 품종(species)을 제외한 4개 변수의 boxplot
sns.boxplot(data=iris)
plt.show()
# DataFrame의 모든 숫자타입 변수들의 boxplot을 하나의 figure 내에 그림
```

## matplotlib의 plt.boxplot() 사용

```python
plt.boxplot(x=[iris[x] for x in iris.columns[:-1]],
           labels=iris.columns[:-1])
plt.show()
```

## 그룹별로 다른 subplot에 boxplot 나타내기

### matplotlib과 sns.boxplot() 혼용

```python
# 품종별로 subplot 컬럼 구분, 데이터프레임 나머지 4개 변수의 boxplot을 품종별로 나타내기
species=iris['species'].unique()
species
```

```python
fig, axes=plt.subplots(ncols=len(species), figsize=(18,6), sharey=True)

for i,s in enumerate(species):
    subset=iris[iris['species']==s]
    sns.boxplot(data=subset, ax=axes[i])
    axes[i].set_title(f'species={s}')

plt.show()
```

### seaborn의 sns.catplot() 사용

```python
sns.catplot(data=iris, col='species', kind='box')
plt.show()
```

# scatter plot

## seaborn의 sns.scatterplot() 사용

```python
# petal.length ~ petal.width 사이 산점도를 품종별로 다른 색의 점으로 나타내기
sns.scatterplot(data=iris, x='petal_width', y='petal_length', hue='species')
plt.show()
```

## matplotlib의 plt.scatter() 사용

```python
setosa=iris[iris['species']=='setosa']
versicolor=iris[iris['species']=='versicolor']
virginica=iris[iris['species']=='virginica']

data=[setosa, versicolor, virginica]
name=['setosa', 'versicolor', 'virginica']

for d,n in zip(data,name):
    plt.scatter(x='petal_width', y='petal_length', data=d, label=n)

plt.legend()
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.show()
```

```python
species=iris['species'].unique()
species
```

```python
for s in species:
    subset=iris[iris['species']==s]
    plt.scatter(x='petal_width', y='petal_length', data=subset, label=s)

plt.legend()
plt.show()
```

# pair plot

## matplotlib 사용 시

```python
def iris_scatter_subplot(data, x, y, ax):
    species=data['species'].unique()
    for s in species:
        subset=data[data['species']==s]
        ax.scatter(x, y, data=subset, label=s)
    ax.legend()
    ax.set_xlabel(x)
    ax.set_ylabel(y)
```

```python
fig, axes=plt.subplots(nrows=4, ncols=4, figsize=(16, 16))

features=iris.columns[:-1]

for row, feature_x in enumerate(features):
    for col, feature_y in enumerate(features):
        iris_scatter_subplot(data=iris, x=feature_y, y=feature_x, ax=axes[row, col])
plt.show()
```

## seaborn의 sns.pairplot() 사용 시
