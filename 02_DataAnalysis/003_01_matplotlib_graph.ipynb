# matplotlib.pyplot을 사용한 배열 시각화

```python
import matplotlib.pyplot as plt
import numpy as np
```

# matplotlib.pyplot 으로 그래프 그리기

```python
x=np.arange(-3,3,1)
print(x)
y=x**2
print(y)
```

```python
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
x=np.arange(-3,3,0.01)
y=x**2
print(x[:10])
print(x[-10:])
plt.plot(x,y, 'r')
plt.grid()
plt.show()
```

```python
x=np.arange(-1.1, 1.1, 0.01)
y=x**3-x
plt.plot(x,y,'b')
plt.grid()
plt.show()
```

```python
x=np.arange(-3,3,0.01)
y=2**x
plt.plot(x,y)
plt.grid()
plt.show()
```

# matplotlib.pyplot에 2개 이상의 그래프를 그리려는 경우

```python
x=np.arange(-3, 3, 0.01)
y=np.exp(x)
y2=2**x
y3=np.log(x)

plt.plot(x,y)
plt.plot(x,y2)
plt.plot(x,y3)
plt.grid()
plt.show()
```

```python
# 지수함수 vs 로그함수
x1=np.arange(-20, 3, 0.01)
y1=np.exp(x1)

x2=np.arange(0.00001, 10, 0.01)
y2=np.log(x2)

x3=np.arange(-10, 10, 0.01)
y3=x3

plt.plot(x1, y1, label='y=e^x')
plt.plot(x2, y2, label='y=log(x)')
plt.plot(x3, y3, label='y=x')

plt.legend()
plt.grid()
plt.axis('square')
plt.show()
```
