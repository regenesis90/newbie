# import

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json # read/write json
```

# Load Data 

* https://github.com/wesm/pydata-book/blob/2nd-edition/datasets/bitly_usagov/example.txt

```python
file_path = 'example.txt'
```

# Open : Read JSON File

```python
with open(file_path, mode = 'r', encoding = 'cp1252') as f:
    records = [json.loads(line) for line in f]
    print(len(records))
```

# Figure Out Data Structure

```python
df = pd.DataFrame(records)
df.head() #> This file has client information accessed usa.gov
```

```python
#> when you change json file(비정형, 반정형), there are many NA in dataframe.
df.info()
```

# How many timezone at 'tz' column?

```python
df['tz']
```

```python
df['tz'].value_counts() #> value_counts() method isn't count NA.
```

```python
df['tz'].value_counts().iloc[:10]
```

# .fillna() method

```python
# if you want count NA, you can use .fillna() method.
df_clean = df.copy() #first, you should copy DataFrame. To avoid transforming original data.
df_clean['tz'] = df_clean['tz'].fillna('Missing') # Replace NA to 'Missing'
```

```python
df_clean['tz'].value_counts().head(10)
```

```python
# if value of tz is ''(empty string), then replace 'Unknown'. else, use original value.
s = df_clean['tz'].value_counts().head(10)
s
```

```python
sns.barplot(x = s.values, y = s.index)
plt.show()
```

# Are there Asia/Seoul in timezone?

```python
df_clean[df_clean['tz'].str.contains('Seoul')]
```

# count of accessed countries

```python
df_clean['c'].value_counts().head(10)
```

```python
df_clean['c'] = df_clean['c'].fillna('Missing')
df_clean['c'].value_counts().head(10)
```

# OS information of accessed client

```python
df_clean['a'] = df_clean['a'].fillna('Missing') # Replace NA to 'Missing' in a column.
df_clean[df_clean['a'].str.contains('Windows')]
```

```python
df_clean[df_clean['a'].str.contains('Linux')]
```

# Add derived variable 'OS' to df_clean DF.

```python
df_clean['OS'] = np.where(df_clean['a'].str.contains('Windows'), 'Windows', 'Other')
df_clean.head()
```

# count by tz, os

## using groupby

```python
df_clean.groupby(by = ['tz', 'OS']).size()
```

```python
df_clean.groupby(by = ['tz','OS']).size().unstack()
```

## using pivot_table()

```python
tz_os_counts = df_clean.pivot_table(index = 'tz', columns = 'OS', aggfunc = 'size')
tz_os_counts
```
